```python
# COPY PSTE THIS CELL

!apt-get install swi-prolog
!pip install pyswip

from pyswip import Prolog

prolog = Prolog()

# Define gender facts (male and female)
prolog.assertz("male(kewal)")
prolog.assertz("male(jayant)")
prolog.assertz("male(dattaraj)")
prolog.assertz("female(mrinmayee)")
prolog.assertz("female(suprabha)")
prolog.assertz("female(ujjawala)")



# Define family relationships (parent facts)
prolog.assertz("parent(jayant, mrinmayee)")    # John is Sarah's parent
prolog.assertz("parent(jayant, kewal)")      # John is Tom's parent
prolog.assertz("parent(suprabha, mrinmayee)")    # Mary is Sarah's parent
prolog.assertz("parent(suprabha, kewal)")      # Mary is Tom's parent
prolog.assertz("parent(dattaraj, suprabha)")    # Sarah is Kate's parent
prolog.assertz("parent(ujjawala, suprabha)")   # Sarah is Lucas's parent
      # Tom is Lily's parent



# Define rules for family relationships

# Sibling rule: X and Y are siblings if they share the same parent
prolog.assertz("sibling(X, Y) :- parent(P, X), parent(P, Y), X \\= Y")

# Grandparent rule: X is a grandparent of Y if X is a parent of Z and Z is a parent of Y
prolog.assertz("grandparent(X, Y) :- parent(X, Z), parent(Z, Y)")

# Mother rule: X is the mother of Y if X is a parent of Y and X is female
prolog.assertz("mother(X, Y) :- parent(X, Y), female(X)")

# Father rule: X is the father of Y if X is a parent of Y and X is male
prolog.assertz("father(X, Y) :- parent(X, Y), male(X)")

# Grandmother rule: X is the grandmother of Y if X is a female and a parent of Z, and Z is a parent of Y
prolog.assertz("grandmother(X, Y) :- mother(X, Z), parent(Z, Y)")

# Grandfather rule: X is the grandfather of Y if X is a male and a parent of Z, and Z is a parent of Y
prolog.assertz("grandfather(X, Y) :- father(X, Z), parent(Z, Y)")


# Use `setof/3` to ensure unique results in queries

# Query 1: Find Sarah's siblings
print("Mrinmayee's siblings:")
siblings = list(prolog.query("setof(Sibling, sibling(mrinmayee, Sibling), SiblingList)"))
if siblings:
    for sibling in siblings[0]['SiblingList']:
        print(sibling)


# Query 2: Find Kate's grandparents
print("\nKewal's grandparents:")
grandparents = list(prolog.query("setof(Grandparent, grandparent(Grandparent, kewal), GrandparentList)"))
if grandparents:
    for grandparent in grandparents[0]['GrandparentList']:
        print(grandparent)



# Query 3: Find Tom's children
print("\nJayant's children:")
children = list(prolog.query("setof(Child, parent(jayant, Child), ChildList)"))
if children:
    for child in children[0]['ChildList']:
        print(child)


# Query 4: Find who is Kate's mother
print("\nSuprabha's mother:")
for result in prolog.query("mother(Mother, suprabha)"):
    print(result['Mother'])


# Query 5: Find who is Kate's father
print("\nSuprabha's father:")
for result in prolog.query("father(Father, suprabha)"):
    print(result['Father'])


# Query 6: Find who is Kate's grandmother
print("\nKewal's grandmother:")
grandmothers = list(prolog.query("setof(Grandmother, grandmother(Grandmother, kewal), GrandmotherList)"))
if grandmothers:
    for grandmother in grandmothers[0]['GrandmotherList']:
        print(grandmother)



# Query 7: Find who is Kate's grandfather
print("\nKewal's grandfather:")
grandfathers = list(prolog.query("setof(Grandfather, grandfather(Grandfather, kewal), GrandfatherList)"))
if grandfathers:
    for grandfather in grandfathers[0]['GrandfatherList']:
        print(grandfather)
```
