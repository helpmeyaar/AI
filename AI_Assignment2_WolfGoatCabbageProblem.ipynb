**2  Implement Constraint Satisfaction Problem.**

```python
# COPY PASTE THIS CELL

def initialize_game():
    return {
        "left": {"farmer", "wolf", "goat", "cabbage"},
        "right": set(),
        "farmer_side": "left"
    }


def is_safe(side):
    return not (
        ("wolf" in side and "goat" in side and "farmer" not in side) or
        ("goat" in side and "cabbage" in side and "farmer" not in side)
    )


def move_item(state, item):
    farmer_side = state["farmer_side"]
    left_side = state["left"]
    right_side = state["right"]

    if farmer_side == "left":
        if item:
            if item not in left_side:
                return "Invalid move!", state
            left_side.remove(item)
            right_side.add(item)
        left_side.remove("farmer")
        right_side.add("farmer")
        state["farmer_side"] = "right"
    else:
        if item:
            if item not in right_side:
                return "Invalid move!", state
            right_side.remove(item)
            left_side.add(item)
        right_side.remove("farmer")
        left_side.add("farmer")
        state["farmer_side"] = "left"

    if not is_safe(left_side) or not is_safe(right_side):
        return "Move not allowed! Items left alone unsafely.", state

    return "Move successful!", state


def display_state(state):
    left_side = ", ".join(state["left"])
    right_side = ", ".join(state["right"])
    farmer_side = state["farmer_side"]
    print(f"Left Side: {left_side}  |  Right Side: {right_side}")
    print(f"Farmer is on the {farmer_side} side.\n")


def is_game_won(state):
    return not state["left"]


def main():
    print("Welcome to the Wolf, Goat, and Cabbage game!\n")
    state = initialize_game()
    display_state(state)

    while not is_game_won(state):
        move = input("What do you want to move? (wolf, goat, cabbage, none): ").strip().lower()

        if move == "none":
            move = None
        elif move not in {"wolf", "goat", "cabbage", None}:
            print("Invalid item! Please choose wolf, goat, cabbage, or none.")
            continue

        result, state = move_item(state, move)
        print(result)
        display_state(state)

        if result.startswith("Move not allowed"):
            break

    if is_game_won(state):
        print("Congratulations! You won the game!")
    else:
        print("Game over! Try again.")



if __name__ == "__main__":
    main()
```
